```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dp=pd.read_csv("/content/sample_data/ModifiesInsurance2.csv")
df=dp.copy()
```

```python
df
```

```python
df['Index']=np.arange(len(df))
```

```python
age_bins = [0, 18, 25, 35, 100]
age_labels = ['Child', 'Young Adult', 'Adult', 'Senior']
df['Age Group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
```

```python
age_bins = [0, 18, 25, 35, 100]
age_labels = ['0-18', '18-25', '25-35', '35-100']
df['Age Interval'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
x=sns.countplot(data=df,x='Age Interval',palette='rocket')
plt.xticks(rotation=60)
for bars in x.containers:
  x.bar_label(bars)
```

The majority of individuals who have obtained insurance are aged 35 years or above

```python
x=sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in x.containers:
  x.bar_label(bars)
```

The insurance policy has been predominantly taken by men.

Insurance trends show that individuals aged 35 and above are the primary policyholders, with a significant representation of men.
